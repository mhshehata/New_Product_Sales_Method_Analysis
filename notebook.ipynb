- nb_sold: no missing data; no cleaning needed.
- revenue: 1074 values are missing. We will fix it using the appropriate method.
- years_as_customer: no missing data; two values were higher than 40, which made no sense since the company was founded 40 years ago. We will replace these values with the max number possible.

## Data Cleaning

- sales_method column: fixing typos and standardize the of writing the three methods as "Email, Call, and Email + Call".
- revenue column: checking the percentage of the missing values; found that it is higher than 5%; filled in the missing values with the 'Imputating with Sub-Group' method using the median of the revenue for each sales_method sub-group.
- years_as_customer column: replaced the unreasonable values with the max value of 40 years.


After data validation, cleaning and re-validation, the dataset contains **15000 rows and 8 columns** without missing values.
